```python
import pandas as pd         #packages
import datetime 
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv("Probe1_Stripling.csv",skiprows= range(1,17))        #Read data and create time column

data.rename(columns={'Time':'Date'}, inplace=True)
data['Date'] = pd.to_datetime(data['Date']) 
data['Time']= data['Date'].dt.time

meteo = pd.read_csv("necessary_meteo_data.csv")

meteo
```

```python
    #Reset index and keep values only between that time period
    
df = data.set_index('Date').between_time('04:45:00', '05:15:00').reset_index().reindex(columns= data.columns)
df['Julian Day'] = df['Date'].dt.dayofyear

df['Date']=df['Date'].dt.date
df = df.drop_duplicates(subset=['Date'],keep= 'first')
df
```

```python
df['avg_sm']= df['sm sum']/6
df
```

```python
df_merged = pd.merge(df,meteo,on='Julian Day')
df_merged.to_excel('Products/probe1_full.xlsx',index=False)
```
